# <font color='black'>Basics TensorFlow</font>

---
We start creating a simple computational graph:

tf.Graph represents a collection of tf.Operations
You can create operations by writing out equations.
By default, there is a graph: tf.get_default_graph()
and any new operations are added to this graph.
The result of a tf.Operation is a tf.Tensor, which holds the values.

```python
x = tf.Variable(3, name = "x")
y = tf.Variable(4, name = "y")
f = x*x*y + y + 2
```

To evaluate this graph, we need to open a TensorFlow session and use it to initialize the variables and evaluate f:

```python
sess = tf.Session()
sess.run(x.initializer)
sess.run(y.initializer)
result = sess.run(f)
print(result)
sess.close()
```

Or equivalently,

```python
with tf.Session() as sess:
    x.initializer.run()
    y.initializer.run()
    result = f.eval()
```

Or, we can create a InteractiveSession instead:

```python
init = tf.global_variables_initializer() # prepare an init node in the graph

sess = tf.InteractiveSession()
init.run() # initialize all the variables
result = f.eval()
print(result)
sess.close()
```

## <font color='black'> Example : Linear Regression with TensorFlow</font>

Firstly, let's create one example

```python
num_samples = 100
xs = np.linspace(-3, 3, num_samples)
ys = np.sin(xs) + np.random.uniform(-0.2, 0.2, num_samples)
plt.scatter(xs, ys)
```

tf.placeholders for the input and output of the network

```python
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)
```

```python
#Define the model parameters, the cost function and the optimizer
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')
Y_pred = tf.add(tf.multiply(X, W), b)

cost = tf.reduce_sum(tf.pow(Y_pred - Y, 2)) / (num_samples - 1)

learning_rate = 0.1
optimizer = tf.train.GradientDescentOptimizer(learning_rate).minimize(cost)
```

```python
# %% We create a session to use the graph
n_epochs = 10
init = tf.global_variables_initializer() # prepare an init node in the graph

with tf.Session() as sess:
    init.run()
    
    training_cost_old = 0.0
    for epoch_i in range(n_epochs):
        for (x, y) in zip(xs, ys):
            sess.run(optimizer, feed_dict={X: x, Y: y})

        training_cost_new = sess.run(cost, feed_dict={X: xs, Y: ys})
        print(training_cost_new)
       
        if np.abs(training_cost_old - training_cost_new) < 0.000001:
            break
        training_cost_old = training_cost_new
        
    plt.scatter(xs, ys)
    plt.plot(xs, Y_pred.eval(feed_dict={X: xs}, session=sess) )  
```
